```python
def data_descritization(d):
        l = d.columns
        for s in l:
            print(type(s))
            print(s)
            if (s=='id' or s=='diagnosis'):
                continue;
            else:
                s1=s+"_labels"
                print(s1)
                print(type(s1))
                d[s1]=pd.cut(d[s],bins=5,labels=[0 , 1 , 2 , 3 , 4])
                del d[s]

                
                
data_descritization(d)
print(d)
```
